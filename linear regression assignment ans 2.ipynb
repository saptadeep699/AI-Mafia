```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# importing necessary data
df_train = pd.read_csv(r'C:\Users\KIIT\Downloads/Train.csv')
print(df_train.shape, type(df_train), df_train, sep='\n\n')
```

```python
# transferring Data into nparray
x1Train = df_train['feature_1'].values.reshape((-1,1))
x2Train = df_train['feature_2'].values.reshape((-1,1))
x3Train = df_train['feature_3'].values.reshape((-1,1))
x4Train = df_train['feature_4'].values.reshape((-1,1))
x5Train = df_train['feature_5'].values.reshape((-1,1))

xTrain=df_train[['feature_1','feature_2','feature_3','feature_4','feature_5']].values
yTrain=df_train['target'].values.reshape((-1,1))

print(x1Train.shape, type(x1Train),x1Train ,x2Train , x3Train ,x4Train ,x5Train, sep='\n\n')
print('\n\n\n',xTrain.shape, type(xTrain),xTrain ,yTrain.shape, type(yTrain), yTrain, sep='\n\n' )
```

```python
# visualising the training data

plt.figure(figsize=(20,20))
plt.scatter(x1Train, yTrain, color='violet', label='feature-1')
plt.scatter(x2Train, yTrain, color='green', label='feature-2')
plt.scatter(x3Train, yTrain, color='orange', label='feature-3')
plt.scatter(x4Train, yTrain, color='grey', label='feature-4')
plt.scatter(x5Train, yTrain, color='yellow', label='feature-5')

plt.title('actual output given in training data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
apmodel= LinearRegression()
```

```python
#training the apmodel by fitting the training data into it
apmodel.fit(xTrain,yTrain)
```

```python
output=apmodel.predict(xTrain)
print(output.shape, type(output), output, sep='\n\n')
```

```python
bias=apmodel.intercept_
coeff=apmodel.coef_
print(bias, coeff, sep='\n\n')
```

```python
apmodel.score(xTrain, yTrain)
```

```python
plt.figure(figsize=(20,20))
plt.scatter(x1Train, yTrain, color='violet', label='training data-feature1')
plt.scatter(x2Train, yTrain, color='green', label='training data-feature2')
plt.scatter(x3Train, yTrain, color='orange', label='training data-feature3')
plt.scatter(x4Train, yTrain, color='grey', label='training data-feature4')
plt.scatter(x5Train, yTrain, color='yellow', label='training data-feature5')

plt.plot(xTrain,output, color='lightblue', label='prediction')
plt.title('actual output vs prediction')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
